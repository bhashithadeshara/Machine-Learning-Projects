# Creation of Basket

## UK Basket 

```python
#filters the "MB1" DataFrame to include only the rows where the "Country" column equals 'United Kingdom',
#groups the resulting DataFrame by both "InvoiceNo" and "Description", sums the "Quantity" values for each group,
#and then pivots the resulting table such that the "Description" values become the columns, and the "Quantity" 
#values become the cell values.

UnitedKingdom = (MB1[MB1['Country'] =="United Kingdom"]
          .groupby(['InvoiceNo', 'Description'])['Quantity']
          .sum().unstack().reset_index().fillna(0)
          .set_index('InvoiceNo'))
```

```python
UnitedKingdom
```

### converting sales data into a binary matrix representation suitable for association rule mining

```python
#This code can be useful for converting sales data into a binary matrix representation suitable for association rule mining.
#In this case, the resulting "UK_basket" DataFrame will have binary values of 0 or 1 depending on whether a particular product
# was purchased on an invoice or not.

#This binary matrix representation can be used to calculate frequent itemsets and generate association rules to understand 
# patterns of co-occurrence among different products purchased by customers in the United Kingdom.

def e(x):
    if x <= 0:
        return 0
    if x >= 1:
        return 1

UK_basket=UnitedKingdom.applymap(e)
```

```python
#view uk_basket
UK_basket
```

```python
#import libraries
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

###  implementing Apriori algorithm for Germay basket

```python
#converts the transactional dataset UK_basket into a boolean format.
#This is required because the Apriori algorithm expects a boolean input where each column represents an item 
#and each row represents a transaction, with a True or False value indicating whether or
#not the item was present in the transaction.
UK_basket_bool = UK_basket.astype(bool)

#applies the Apriori algorithm to the boolean dataset. 
#The min_support parameter sets the minimum support threshold for an itemset to be considered frequent, 
#while the use_colnames parameter specifies that the resulting DataFrame should use the item names as column labels 
#instead of the default column indices.
freq_itemsets = apriori(UK_basket_bool, min_support=0.025, use_colnames=True)
```

```python
#View frequent itemsets
freq_itemsets
```

### Generate association rules from frequent itemsets

```python
#The metric parameter specifies the metric to be used to evaluate the association rules. 
#In this case, the metric is set to "confidence", which measures the conditional probability of the consequent given
#the antecedent.

#The min_threshold parameter sets the minimum threshold for the chosen metric that an association rule must meet in order
#to be considered interesting. In this case, the threshold is set to 0.7, meaning that only rules with 
#a confidence score of 0.7 or higher will be returned.
association_rules(freq_itemsets, metric="confidence", min_threshold=0.7)
```

### applies the association_rules() function to the frequent itemsets

```python
rules1 = association_rules(freq_itemsets, metric="lift", min_threshold=10)
rules1.head()
```

### applying a lambda function to the "antecedents" column of the DataFrame using the apply() function

```python
#The lambda function simply calculates the length of each antecedent set in the "antecedents" column using the len() function.
#This results in a new column that contains the length of each antecedent set for each association rule in the DataFrame.

#This new column can be useful for analyzing the relationship between the length of the antecedent set 
#and the performance of the association rule, such as its lift or confidence scores. 
#It can also help identify any patterns or trends in the data that might be missed by simply looking at the raw statistics.

rules1["antecedent_len"] = rules1["antecedents"].apply(lambda x: len(x))
```

###  filters the rules1 DataFrame to include only the association rules 

```python
rule1=rules1[ (rules1['antecedent_len'] >= 1) &
       (rules1['confidence'] > 0.7) &
       (rules1['lift'] > 10) ]
```

### filters the rule1 DataFrame to only include rules that have at least one antecedent item

```python
#By sorting the rules by lift score in descending order, 
#identify the rules that have the strongest positive associations between the antecedent and consequent items. 
#These rules can be especially useful for generating insights and identifying potential cross-selling opportunities.

rule1[rule1['antecedent_len']>=1].sort_values('lift', ascending=False).head()
```

```python
import seaborn as sns
#To convert the frozen sets with strings we add a column
rule1['antecedents_'] = rule1['antecedents'].apply(lambda a: ','.join(list(a)))
rule1['consequents_'] = rule1['consequents'].apply(lambda a: ','.join(list(a)))

# using the lift metric transform the DataFrame of rules into a matrix 
pivot1 = rule1[rule1['antecedent_len']>0].pivot(index = 'antecedents_', 
                    columns = 'consequents_', values= 'lift')

# Generate a heatmap with annotations on and the colorbar off

sns.heatmap(pivot1, annot = True)
plt.show(sns)
```
